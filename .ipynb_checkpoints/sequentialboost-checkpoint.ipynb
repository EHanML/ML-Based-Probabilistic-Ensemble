```python
# NOIS
def bool2int(bool):
    boolint = np.full(fill_value=1,shape=len(bool))
    boolint[~bool] =0
    return boolint
    
def get_NOIS(y_true,y_pred,y_l,y_u,alpha):
    awpi = np.mean(y_u - y_l)
    nois = awpi+ (2/alpha)*np.mean(np.multiply((y_l- y_true), bool2int(y_l > y_true)) + 
                          np.multiply((y_true - y_u), bool2int(y_true > y_u))  )
    return nois
```

```python
def LGB_run(X_train, Y_train, X_dev, Y_dev, params, verbose = 100):
    
    train = lgb.Dataset(data=X_train,
                        label=Y_train,
                        feature_name = predictors,
                        # feature_name='auto'
                        group=None, categorical_feature=['E_Label_ParcelID'])
    
    dev = lgb.Dataset(data=X_dev,
                        label=Y_dev,
                        reference=train,
                        feature_name = predictors,
                      # feature_name='auto',
                        group=None, 
                        categorical_feature=['E_Label_ParcelID'])
    
    #sklearn
    model = lgb.train(params = params,
                      train_set = train,
                      num_boost_round = 1000,
                      valid_sets = [dev],
                      valid_names = ['eval'],
                      early_stopping_rounds = 100,
                      verbose_eval = verbose)
    
    score = model.best_score['eval']['rmse']
    
    ntree =  model.best_iteration
    
    return model, ntree, score
```

```python
def lgbBayesVal(X_train, Y_train, X_dev, Y_dev, space_params, FUN, max_eval = 10, seed = 1):
    
    def objective(params):
        params = {   
            'objective': 'regression',
            #'reg_sqrt': True, 
            'boosting': 'gbdt',
            'boost_from_average': True,
            'metric':'rmse',
            'tree_learner': 'serial',
            'nthread': 8, # for the best speed, set this to the number of real CPU cores, not the number of thread
            'device_type': 'cpu',
            'seed': seed,
            'eta': 0.03,   
            #'max_bin': 500,
            'num_leaves': int(params['num_leaves']),
            'max_depth': int(params['max_depth']),
            'min_data_in_leaf': int(params['min_data_in_leaf']),
            'bagging_fraction': "{:.3f}".format(params['bagging_fraction']),
            'feature_fraction': "{:.3f}".format(params['feature_fraction'])
        }
        
        lgbfit, ntree, score  = FUN(X_train, Y_train, X_dev, Y_dev, params, verbose = False)
        print(f'Pars = {params} \t ntree = {ntree} \t RMSE = {score}')
        return {
            'loss': score,
            'status': STATUS_OK,
            # other results
            'ntree': ntree
        }
    
    trials = Trials()
    best = fmin(fn=objective, # function of space, get pars using space['par']
            space=space_params,
            algo=tpe.suggest,
            trials = trials,
            rstate=np.random.RandomState(seed),
            max_evals=max_eval)
    
    
    best_params = space_eval(space_params, best) # this returns dictionary parameter 
    #best_params = trials.argmin # this returns space parameter 
    return best_params, trials.best_trial.get('result').get('ntree')
```

# Oridnary Gradient

def obj_theta2(preds, train_data):
    y = train_data.get_label()
    miu =  lgb_model1.predict(train_data.get_data())
    grad = 0.5 * (preds**-1) - 0.5 * (preds**-2) * ((y - miu)**2)
    hess = -0.5*(preds**-2) + (preds**-3)*((y - miu)**2)
    return grad, hess

def eval_NNL(preds, train_data):
    y = train_data.get_label()
    miu =  lgb_model1.predict(train_data.get_data())
    pi =  np.full(fill_value= np.pi, shape=np.shape(y)[0],dtype=float)
    nll = 0.5 * np.sum(np.log(2*pi) + np.log(preds) + (preds**-1) * ((y - miu)**2))
    return 'NLL', nll, False
    

```python
def obj_logtheta(preds, train_data):
    y = train_data.get_label()
    miu =  lgb_model1_dev.predict(train_data.get_data())
    grad = 1 - np.exp(-2*preds) * ((y - miu)**2)
    hess =  2 * np.exp(-2*preds) * ((y - miu)**2)
    return grad, hess

def eval_NNL(preds, train_data):
    y = train_data.get_label()
    miu =  lgb_model1_dev.predict(train_data.get_data())
    pi =  np.full(fill_value= np.pi, shape=np.shape(y)[0],dtype=float)
    nll = (0.5 * np.sum(np.log(2*pi) + preds + 0.5 * ((y - miu)**2) * np.exp(-2*preds)))/y.shape[0]
    return 'NLL', nll, False
```
