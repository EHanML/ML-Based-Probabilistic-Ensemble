# XGB

```python
def XGB_run(X_train, Y_train, X_dev, Y_dev, params, verbose = 100):
    
    dtrain = xgb.DMatrix(X_train, Y_train)
    ddev = xgb.DMatrix(X_dev, Y_dev)
    watchlist = [(dtrain, 'train'), (ddev, 'eval')]
    
    model = xgb.train(params = params,
                      dtrain = dtrain,
                      num_boost_round = 1000,
                      evals = watchlist,
                      maximize=False,
                      early_stopping_rounds = 100,
                      verbose_eval=verbose)
    return model, model.best_ntree_limit, model.best_score
```

```python
# try
param = {'max_depth':2, 'eta':0.01, 'silent':1, 'objective':'reg:squarederror'}
xgbfit, ntree, score  = XGB_run(X_train, Y_train, X_dev, Y_dev, params = param)
print(ntree, score )
```

```python
from hyperopt import fmin, tpe, space_eval, hp, rand, Trials, STATUS_OK
from hyperopt.pyll.stochastic import sample
```

```python
def xgbBayesVal(X_train, Y_train, X_dev, Y_dev, space_params, FUN, max_eval = 10, seed = 1):
    
    def objective(params):
        params = {   
            'booster': 'gbtree',
            'nthread': 10,
            'eta': 0.1,
            'tree_method': 'auto',
            'max_depth': int(params['max_depth']),
            'gamma': "{:.3f}".format(params['gamma']),
            'min_child_weight': "{:.3f}".format(params['min_child_weight']),
            'subsample': "{:.3f}".format(params['subsample']),
            'colsample_bytree': "{:.3f}".format(params['colsample_bytree'])
        }
        xgbfit, ntree, score  = FUN(X_train, Y_train, X_dev, Y_dev, params, verbose = False)
        print(f'Pars = {params} \t ntree = {ntree} \t RMSE = {score}')
        return {
            'loss': score,
            'status': STATUS_OK,
            # other results
            'ntree': ntree
        }
    
    trials = Trials()
    best = fmin(fn=objective, # function of space, get pars using space['par']
            space=space_params,
            algo=tpe.suggest,
            trials = trials,
            rstate=np.random.RandomState(seed),
            max_evals=max_eval)
    
    
    best_params = space_eval(space_params, best) # this returns dictionary parameter 
    #best_params = trials.argmin # this returns space parameter 
    return best_params, trials.best_trial.get('result').get('ntree')
```

```python
space_params = {   
    'max_depth': hp.quniform('max_depth', 2, 12, 1),
    'gamma': hp.uniform('gamma', 0, .7),
    'min_child_weight': hp.loguniform('min_child_weight', 0, 5),
    'subsample': hp.uniform('subsample', 0.5, 1),
    'colsample_bytree': hp.uniform('colsample_bytree', 0.5, 1)   
}
```

```python
best_par, ntree = xgbBayesVal(X_train, Y_train, X_dev, Y_dev, space_params, FUN = XGB_run, max_eval=100)
```

```python
# best par
print(best_par, ntree)
```

# LGB

```python
def LGB_run(X_train, Y_train, X_dev, Y_dev, params, verbose = 100):
    
    train = lgb.Dataset(data=X_train,
                        label=Y_train,
                        feature_name = predictors,
                        # feature_name='auto'
                        group=None, categorical_feature='auto')
    
    dev = lgb.Dataset(data=X_dev,
                        label=Y_dev,
                        reference=train,
                        feature_name = predictors,
                      # feature_name='auto',
                        group=None, 
                        categorical_feature='auto')
    
    #sklearn
    model = lgb.train(params = params,
                      train_set = train,
                      num_boost_round = 1000,
                      valid_sets = [dev],
                      valid_names = ['eval'],
                      early_stopping_rounds = 100,
                      verbose_eval = verbose)
    
    score = model.best_score['eval']['rmse']
    
    ntree =  model.best_iteration
    
    return model, ntree, score
```

```python
def lgbBayesVal(X_train, Y_train, X_dev, Y_dev, space_params, FUN, max_eval = 10, seed = 1):
    
    def objective(params):
        params = {   
            'objective': 'regression',
            'boosting': 'gbdt',
            'metric':'rmse',
            'tree_learner': 'serial',
            'num_thread': 8,  # for the best speed, set this to the number of real CPU cores, not the number of thread
            'device_type': 'cpu',
            'seed': seed,
            'eta': 0.1,   
            #'max_bin': 500,
            'num_leaves': int(params['num_leaves']),
            'max_depth': int(params['max_depth']),
            'min_data_in_leaf': int(params['min_data_in_leaf']),
            'bagging_fraction': "{:.3f}".format(params['bagging_fraction']),
            'feature_fraction': "{:.3f}".format(params['feature_fraction'])
        }
        
        lgbfit, ntree, score  = FUN(X_train, Y_train, X_dev, Y_dev, params, verbose = False)
        print(f'Pars = {params} \t ntree = {ntree} \t RMSE = {score}')
        return {
            'loss': score,
            'status': STATUS_OK,
            # other results
            'ntree': ntree
        }
    
    trials = Trials()
    best = fmin(fn=objective, # function of space, get pars using space['par']
            space=space_params,
            algo=tpe.suggest,
            trials = trials,
            rstate=np.random.RandomState(seed),
            max_evals=max_eval)
    
    
    best_params = space_eval(space_params, best) # this returns dictionary parameter 
    #best_params = trials.argmin # this returns space parameter 
    return best_params, trials.best_trial.get('result').get('ntree')
```

```python
space_params = {   
    'num_leaves': hp.qloguniform('num_leaves', 3, 11, 1),
    'max_depth': hp.choice('max_depth', [-1, 6, 10, 20, 30, 50]),
    'min_data_in_leaf': hp.qloguniform('min_data_in_leaf', 2, 8, 1),
    'bagging_fraction': hp.uniform('bagging_fraction', 0.5, 1),
    'feature_fraction': hp.uniform('feature_fraction', 0.5, 1)   
}
```

```python
best_par, ntree = lgbBayesVal(X_train, Y_train, X_dev, Y_dev, space_params, FUN = LGB_run, max_eval=100)
```

```python
# best par
print(best_par, ntree)
```
